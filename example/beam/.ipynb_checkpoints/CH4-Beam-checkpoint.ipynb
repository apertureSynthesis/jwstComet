```python
def jy_2_wn(wave,spec,dspec,title):
    wavem = [i*u.um for i in wave]
    waveh = [i.to(u.Hz,equivalencies=u.spectral()) for i in wavem]
    wavek = [10000 / i / u.cm for i in wave]

    flux = spec*u.Jy
    nflux = flux.to(u.W/u.m**2/u.Hz)
    nu_f_nu = [i*j for i,j in zip(nflux,waveh)]
    flux_k = [i/j for i,j in zip(nu_f_nu,wavek)]

    err_k = [i * j/k for i,j,k in zip(flux_k,dspec,spec)]

    wavef = [i.value for i in wavek]
    fluxf = [i.value for i in flux_k]
    errf = [i.value for i in err_k]

    fspec = np.vstack((wavef,fluxf,errf)).T
    np.savetxt(title,fspec)
    return wavef, fluxf, errf
    
```

```python
ch4_spec = spec - base - h2o_model(wave) - ohp_model(wave) - ch3d_model(wave)
ch3d_spec = spec - base - h2o_model(wave) - ohp_model(wave) - ch4_model(wave)
ohp_spec = spec - base - h2o_model(wave) - ch4_model(wave) - ch3d_model(wave)
```
